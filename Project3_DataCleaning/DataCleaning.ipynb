```python
import pandas as pd
import numpy as np
df = pd.read_csv('AB_NYC_2019.csv')
```

```python
print(f"Dataset: {df.shape}")
```

```python
df.head()
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
no_reviews = df[df['last_review'].isnull()]
print(f"Listings with no reviews: {len(no_reviews)}")
print(f"Their number_of_reviews: {no_reviews['number_of_reviews'].unique()}")
```

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
print(df['reviews_per_month'].isnull().sum())
```

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df['name'] = df['name'].fillna('Unknown')
df['host_name'] = df['host_name'].fillna('Unknown')

print(df.isnull().sum())
```

```python
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
print(df['last_review'].dtype)
```

```python
print("Price stats:")
print(df['price'].describe())
print(f"\nListings with $0 price: {(df['price'] == 0).sum()}")
print(f"Listings with price > $10,000: {(df['price'] > 10000).sum()}")
```

```python
print("Minimum nights stats:")
print(df['minimum_nights'].describe())
print(f"\nListings requiring 365+ nights: {(df['minimum_nights'] >= 365).sum()}")
```

```python
df = df[df['price'] > 0]
df = df[df['minimum_nights'] < 365]
```

```python
print(f"Rows after cleaning: {len(df)}")
print(f"Removed: {48895 - len(df)} rows")
```

```python

print("DATA CHECK")

print(f"\nTotal rows: {len(df)}")
print(f"\nMissing values:")
print(df.isnull().sum())

print(f"\nPrice range: ${df['price'].min()} - ${df['price'].max()}")
print(f"Minimum nights range: {df['minimum_nights'].min()} - {df['minimum_nights'].max()}")

print("\n Ready for use")
```

```python
df.to_csv('AB_NYC_2019_cleaned.csv', index=False)
```
